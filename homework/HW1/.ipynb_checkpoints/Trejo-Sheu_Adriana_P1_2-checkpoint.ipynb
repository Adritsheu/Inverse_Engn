## Problem 1: What Polynomial Did we Pick? 
### Description
Find the coefficients of the 10th order polynomial. Find the estimated errors in the coefficients

No intercept term

**note** used a np.random to create noise \
Liyuan Recommends to use Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import MinMaxScaler 
import sklearn.metrics 
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split, cross_validate
```

```python
data = np.loadtxt('problem1_fin.txt')
x,y = [],[]
for i in data:
    x.append(i[0])
    y.append(i[1])
x = np.array(x)
y = np.array(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2, random_state = 1)
```

```python
scale = MinMaxScaler()
x_train = scale.fit_transform(xtrain.reshape(-1,1))
x_test = scale.transform(xtest.reshape(-1,1))
```

```python
ten_poly = PolynomialFeatures(degree = 10, include_bias = False)
xtrain_poly = ten_poly.fit_transform(x_train)
xtest_poly = ten_poly.transform(x_test)
```

```python
lr = LinearRegression(fit_intercept = False) # either need polynomial features of LR to not include bias
lr.fit(xtrain_poly,ytrain)
ypred_train = lr.predict(xtrain_poly)
ypred_test = lr.predict(xtest_poly)
```

```python
mse_train = mse(ytrain, ypred_train)
mse_test = mse(ytest, ypred_test)
print(f"MSE from train data {mse_train}\nMSE from test data {mse_test}")
```

```python
# no intercept term
print("Answer: ")
coeff = lr.coef_
print("Coefficients of the polynomial:", coeff)
```

```python
fx = lambda x: coeff[0]*x + coeff[1]*x**2 + coeff[2]*x**3 + coeff[3]*x**4+coeff[4]*x**5+ coeff[5]*x**6+coeff[6]*x**7 + \
    coeff[7]*x**8 + coeff[8]*x**9 + coeff[9]*x**10

x_list = np.linspace(0,1,1000)
y = [fx(x) for x in x_list]

plt.figure(figsize = (7,5))
plt.title("Showing a Plot of the Function")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xtrain, ytrain, "o",alpha = 0.3, label = "Train data")
plt.plot(xtest, ytest, "*",alpha = 0.3, label = "Test data")
plt.plot(x_list,y, "k--" ,label = "Function using coefficients")
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
ols = sm.OLS(ytrain, xtrain_poly)
ols_result = ols.fit()
ols_result.summary()
```

### Testing Ridge

```python
# Taken from Section1_Regression_Deblurring 
def loss_ridge(ypred,y_real,alpha,beta):
    n = len(ypred) # number of training examples
    return 1/n * ( (np.linalg.norm(ypred-y_real))**2 + alpha*(np.linalg.norm(beta))**2 )
```

```python
alpha_list = [10**-5,10**-4, 10**-3, 10**-2, 10**-1, 1, 10]
deg_try = np.linspace(1,20,20)


loss_dict = {}

for a in alpha_list:
    deg_loss = []
    for deg in deg_try:
        
        # Polynomial Features
        poly = PolynomialFeatures(degree= int(deg), include_bias = False)
        xtrain_poly = poly.fit_transform(x_train)
        xtest_poly = poly.transform(x_test)
        
        # Do ridge fit
        ridge = Ridge(alpha = a)
        ridge.fit(xtrain_poly, ytrain)
        coef = ridge.coef_
        ypred = ridge.predict(xtest_poly) 
        
        # get loss
        loss = loss_ridge(ypred, ytest, a, coef)
        deg_loss.append(loss)
    loss_dict[f"alpha {a}"] = deg_loss
```

```python
def report_values(dict_name):
    ''' Function to report lowest loss for alpha/degree combo
     Input: Name of dictionary
     outputs: Prints values, gives lowest degree & lowest value
     
    '''
    keys = []
    for i in dict_name.keys():
        keys.append(i)

    min_val = 100
    for name in keys:
        list_val = dict_name[name]
        min_list = min(list_val)

        if min_list < min_val:
            low_alpha = name
            low_deg = np.argmin(list_val) + 1
            min_val = min_list

    print(f"Chosen alpha: {low_alpha}, chosen degree: {low_deg}")
    print(f"The resulting Loss value: {min_val}")
    return low_deg, min_val
```

```python
low_deg, min_val = report_values(loss_dict)
```

```python
# Now refit
poly = PolynomialFeatures(degree= low_deg, include_bias = False)
xtrain_poly = poly.fit_transform(x_train)
xtest_poly = poly.transform(x_test)

# Do ridge fit
ridge = Ridge(alpha = 10**-5)
ridge.fit(xtrain_poly, ytrain)
coef_ridge = ridge.coef_
int_ridge = ridge.intercept_
ypred = ridge.predict(xtest_poly) 
print("List of Coefficients:", coef_ridge)
print("Intercept:", int_ridge)
```

```python
fx2 = lambda x: int_ridge + coef_ridge[0]*x + coef_ridge[1]*x**2 + coef_ridge[2]*x**3 + coef_ridge[3]*x**4 + coef_ridge[4]*x**5 + \
coef_ridge[5]*x**6 + coef_ridge[6]*x**7 + coef_ridge[7]*x**8 + coef_ridge[8]*x**9 + coef_ridge[9]*x**10 + \
coef_ridge[10]*x**11 + coef_ridge[11]*x**12 + coef_ridge[12]*x**13 + coef_ridge[13]*x**14 + coef_ridge[14]*x**15 + \
coef_ridge[15]*x**16 + coef_ridge[16]*x**17 + coef_ridge[17]*x**18 + coef_ridge[18]*x**19 + coef_ridge[19]*x**20


x_list = np.linspace(0,1,1000)
y2 = [fx2(x) for x in x_list]
```

```python
#Plot
plt.figure(figsize = (7,5))
plt.title("Plot of test data with Predictions using Ridge Regularization")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xtrain, ytrain, "*",alpha = 0.3, label = "Train data")
plt.plot(xtest, ypred, "o", alpha = 0.3 ,label = "Prediction data")
plt.plot(x_list, y2, "k--" ,label = "Function using coefficients")
plt.legend()
plt.show()
```
